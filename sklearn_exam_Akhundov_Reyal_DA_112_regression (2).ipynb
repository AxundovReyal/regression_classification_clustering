```python
import pandas as pd
import numpy as np
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

df_cars_20_percent = df_cars_20_percent.copy()

imputer = SimpleImputer(strategy='most_frequent')

df_cars_20_percent[['Sahiblər', 'Yerlərin sayı','Hansı bazar üçün yığılıb']] = imputer.fit_transform(df_cars_20_percent[['Sahiblər', 'Yerlərin sayı','Hansı bazar üçün yığılıb']])

X = df_cars_20_percent.drop("price_x", axis=1)
y = df_cars_20_percent["price_x"]

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

num_features = X_train.select_dtypes(include=[np.number]).columns
cat_features = X_train.select_dtypes(exclude=[np.number]).columns

num_pipeline = Pipeline([
    ('impute', SimpleImputer(strategy='median')),
    ('scale', StandardScaler())
])

cat_pipeline = Pipeline([
    ('impute', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', num_pipeline, num_features),
        ('cat', cat_pipeline, cat_features)
    ])

estimator = LinearRegression()

full_pipeline_with_estimator = Pipeline([
    ('preprocessor', preprocessor),
    ('estimator', estimator)
])

full_pipeline_with_estimator.fit(X_train, y_train)
```

```python
import lightgbm as lgb
from sklearn.model_selection import RandomizedSearchCV, cross_val_score
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
from scipy.stats import uniform, randint

# LightGBM modelini yaradın
lgb_estimator = lgb.LGBMRegressor()

# Pipeline qururuq
lgb_pipeline = Pipeline([
    ('preprocessor', preprocessor),
    ('estimator', lgb_estimator)
])

# RandomizedSearch üçün hiperparametr aralığı
lgb_param_dist = {
    'estimator__learning_rate': uniform(0.01, 0.3),  # 0.01 ilə 0.31 arasında rastgele
    'estimator__max_depth': randint(3, 7),  # 3-6 arası təsadüfi dəyərlər
    'estimator__n_estimators': randint(100, 500),  # 100-500 arası təsadüfi dəyərlər
    'estimator__num_leaves': randint(20, 100),  # 20-100 arası təsadüfi dəyərlər
    'estimator__min_child_samples': randint(5, 100),  # 5-100 arası təsadüfi dəyərlər
}

# RandomizedSearchCV tətbiq edirik
lgb_random_search = RandomizedSearchCV(
    lgb_pipeline, lgb_param_dist, n_iter=5, cv=5, scoring='neg_root_mean_squared_error', n_jobs=-1, random_state=42
)
lgb_random_search.fit(X_train, y_train)

# Ən yaxşı parametrləri çap edirik
print(f"LightGBM Optimal Parametrlər: {lgb_random_search.best_params_}")

# Ən yaxşı modeli seçirik
best_lgb_model = lgb_random_search.best_estimator_

# Cross-validation nəticələri
lgb_r2_scores = cross_val_score(best_lgb_model, X_train, y_train, cv=5, scoring='r2', n_jobs=-1)
lgb_rmse_scores = cross_val_score(best_lgb_model, X_train, y_train, cv=5, scoring='neg_root_mean_squared_error', n_jobs=-1)

# Nəticələri çap edirik
print(f"LightGBM Ortalama R²: {np.mean(lgb_r2_scores):.4f}")
print(f"LightGBM Ortalama RMSE: {-np.mean(lgb_rmse_scores):.4f}")

# Test dəstində proqnozlaşdırma
lgb_y_pred = best_lgb_model.predict(X_test)

# LightGBM üçün qiymətləndirmə
lgb_r2_score_val = r2_score(y_test, lgb_y_pred)
lgb_rmse_score_val = np.sqrt(mean_squared_error(y_test, lgb_y_pred))

print(f"LightGBM Test Setindəki R²: {lgb_r2_score_val:.4f}")
print(f"LightGBM Test Setindəki RMSE: {lgb_rmse_score_val:.4f}")
```

```python
import lightgbm as lgb
from sklearn.preprocessing import PolynomialFeatures, StandardScaler
from sklearn.pipeline import Pipeline
from sklearn.model_selection import RandomizedSearchCV
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np

def remove_outliers(X, y, threshold=3.0):

    scaler = StandardScaler()
    # Get numerical features from preprocessor
    num_features = preprocessor.transformers_[0][2]
    # Select only numerical features for scaling
    X_num = X[num_features]
    X_scaled = scaler.fit_transform(X_num)

    # Z-skoru 3-dən böyük olanları outlier kimi təyin edirik
    mask = np.abs(X_scaled) < threshold
    return X[mask.all(axis=1)], y[mask.all(axis=1)]

# PolynomialFeatures tətbiq etmək üçün
poly = PolynomialFeatures(degree=2, include_bias=False)

# LightGBM modelini təyin edirik
lgb_estimator = lgb.LGBMRegressor()


full_pipeline_with_estimator = Pipeline([
    ('preprocessor', preprocessor),  # Apply preprocessor first
    ('poly', poly),  # PolynomialFeatures əlavə edirik
    ('estimator', lgb_estimator)
])



param_distributions = {
    'estimator__learning_rate': [0.01, 0.05, 0.1, 0.2],
    'estimator__max_depth': [3, 4, 5, 6],
    'estimator__n_estimators': [100, 200, 300, 400],
    'estimator__num_leaves': [20, 40, 60, 80],
    'estimator__min_child_samples': [5, 10, 20, 30]
}


X_train = X_train.sample(frac=0.2, random_state=42)
y_train = y_train.sample(frac=0.2, random_state=42)
X_test = X_test.sample(frac=0.2, random_state=42)
y_test = y_test.sample(frac=0.2, random_state=42)

X_train_no_outliers, y_train_no_outliers = remove_outliers(X_train, y_train.values, threshold=3.0)
X_test_no_outliers, y_test_no_outliers = remove_outliers(X_test, y_test.values, threshold=3.0)

random_search = RandomizedSearchCV(full_pipeline_with_estimator, param_distributions, n_iter=5, cv=5, scoring='neg_root_mean_squared_error', n_jobs=-1, random_state=42)
random_search.fit(X_train_no_outliers, y_train_no_outliers)


print(f"Optimal Parametrlər: {random_search.best_params_}")

best_model = random_search.best_estimator_


y_pred = best_model.predict(X_test_no_outliers)

r2_score_val = r2_score(y_test_no_outliers, y_pred)
rmse_score_val = np.sqrt(mean_squared_error(y_test_no_outliers, y_pred))

# Nəticələri çap edirik
print(f"Test Setindəki R²: {r2_score_val:.4f}")
print(f"Test Setindəki RMSE: {rmse_score_val:.4f}")
```
